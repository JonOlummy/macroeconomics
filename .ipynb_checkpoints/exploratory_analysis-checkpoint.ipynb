```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from datetime import datetime
import numpy as np

import warnings
warnings.simplefilter("ignore")
```

```python
# df = pd.read_csv('data/cleaned_data.csv', parse_dates=['Date'])
df = pd.read_csv('data/data.csv', parse_dates=['Date'])
df.columns = df.columns.str.strip()
```

```python
df.info()
```

```python
df
```

```python
df.describe()
```

```python
df.set_index('Date', inplace=True)

for column in df.columns:
    plt.figure(figsize=(14, 6))
    plt.plot(df.index, df[column], label=column)
    plt.title(f'{column}')
    plt.xlabel('Date')
    plt.ylabel(column)
    plt.legend()
    plt.show()
```

```python
# select only numeric columns for the correlation matrix
numeric_data = df.select_dtypes(include=[np.number])

correlation_matrix = numeric_data.corr()

correlation_matrix
```

```python
# Columns relating to Japan
japan_columns = [
    'General government debt, % of GDP: net JPY',
    'General government balance, % of GDP: overall JPY',
    'Main producer price index: %oya, 3mma Japan',
    'Merchandise trade balance ratio change (sa): 3M/3M Japan',
    'Generic government bond returns: 5-year maturity Japan'
]

# Columns relating to UK
uk_columns = [
    'General government debt, % of GDP: net GBP',
    'General government balance, % of GDP: overall GBP',
    'Main producer price index: %oya, 3mma GBP',
    'Merchandise trade balance ratio change (sa): 3M/3M GBP',
    'Generic government bond returns: 5-year maturity GBP'
]

# Japan and UK DataFrames
df_japan = df[japan_columns]
df_uk = df[uk_columns]
```

```python
# Correlation matrix for Japan
correlation_matrix_japan = df_japan.corr()

# Correlation matrix for the UK
correlation_matrix_uk = df_uk.corr()

# Display correlation matrices
print("Correlation Matrix for Japan:")
correlation_matrix_japan
```

```python
print("\nCorrelation Matrix for the UK:")
correlation_matrix_uk
```

```python
# Plot correlation matrix for Japan
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_japan, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Japan')
plt.show()
```

```python
# Plot correlation matrix for the UK
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_uk, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for the UK')
plt.show()
```

```python
columns_of_interest = [
    'General government debt, % of GDP: net GBP',
    'General government balance, % of GDP: overall GBP',
    'Main producer price index: %oya, 3mma GBP',
    'Merchandise trade balance ratio change (sa): 3M/3M GBP'
]

for col in columns_of_interest:
    if 'bond_returns' not in col:
        plt.figure(figsize=(8, 5))
        sns.scatterplot(x=df[col], y=df['Generic government bond returns: 5-year maturity GBP'])
        plt.title(f'Relationship between {col} and Government Bond Returns (GBP)')
        plt.xlabel(col)
        plt.ylabel('Government Bond Returns (GBP)')
        plt.show()
```
